# 문제 1. 미분 -- 10점

함수 $f(x)=\cos(x)e^{-0.1x}$ 를 입력하면 도함수를 리턴하는 함수를 파이썬으로 구현하라. 이러한 함수를 이용하여 $f(x)$, $f'(x)$, $f''(x)$의 그래프를 한 Figure에 모두 그려라. (범례는 표현하지 않아도 감점없음)

- $x$의 범위는 $(-2,15)$ 로 할 것.
- tip: 욕심부려서 $h$의 값을 너무 작게 설정하면 그림이 예쁘게 안나옵니다.

(풀이)

````python
def f(x):
    return np.cos(x)*np.exp(-0.1*x)
x = np.linspace(-2,15,1000)
def d(f):
    return lambda x: (f(x)-f(x+0.01))/0.01
f1 = d(f)
f2 = d(f1)
f3 = d(f2)
plt.plot(x,f(x))
plt.plot(x,f1(x))
plt.plot(x,f2(x))
plt.plot(x,f3(x))
````

# 문제 6. -- 30점



`(1)` 함수 $g(x) = \cos x$라 할 때, $1 \leq x \leq 10$에서 부등식 -- 10점

$$g(x) > \frac{\sqrt{3}}{2}$$

을 만족시키는 모든 자연수 $x$의 개수를 구하시오.

(풀이)

`(2)` 함수 $h(x) = e^{-0.5x}$라 할 때, $1 \leq x \leq 10$에서 부등식 -- 10점

$$h(x) \cdot h(x+1) < 0.001$$

을 만족시키는 모든 자연수 $x$의 값의 곱을 구하시오.

(풀이)

````python
h = lambda z: np.exp(-0.5*z)
(x[h(x) * h(x+1) < 0.001]).prod()
````
