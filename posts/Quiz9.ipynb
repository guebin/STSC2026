# 1. `Repeat` 클래스

주어진 값을 `n`번 반복하여 반환하는 이터레이터 클래스 `Repeat`를 작성하라.

```python
for x in Repeat("안녕", 3):
    print(x)
```
```
안녕
안녕
안녕
```

# 2. `Countdown` 클래스

`n`부터 1까지 카운트다운하는 이터레이터 클래스 `Countdown`를 작성하라.

```python
for x in Countdown(5):
    print(x)
```
```
5
4
3
2
1
```

# 3. `UpDown` 클래스

1부터 `n`까지 올라갔다가 다시 내려오는 이터레이터 클래스 `UpDown`을 작성하라.

```python
for x in UpDown(3):
    print(x)
```
```
1 2 3 2 1
```
