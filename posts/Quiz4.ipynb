# 문제 5. 람다 함수 -- 10점

> 각 소문항 5점씩

`(1)` 다음 파일 이름 리스트가 주어졌을 때:

```python
filenames = ['Report.txt', 'data.csv', 'Image.png', 'script.py', 'Document.pdf']
```

람다 함수를 작성하여 대문자로 시작하는 파일은 'important', 소문자로 시작하는 파일은 'normal'을 반환하게 하라. 그리고 리스트 컴프리헨션을 사용하여 각 파일의 분류를 구하라.

**예상 결과:**
```python
['important', 'normal', 'important', 'normal', 'important']
```

(풀이)

````python
filenames = ['Report.txt', 'data.csv', 'Image.png', 'script.py', 'Document.pdf']
f = lambda x: 'important' if x.isupper() else 'nornmal'
[f(i[0]) for i in filenames]
````

`(2)` 다음 파일 이름 리스트가 주어졌을 때:

**예상 결과:**
```python
['data', 'report', 'image', 'script']
```

(풀이)

````python
filenames = ['data_v1.csv', 'report_v2.txt', 'image_v3.png', 'script_v1.py']
[i.split("_")[0] for i in filenames]
````
