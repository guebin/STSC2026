# 2. 상속 -- 50점

아래와 같은 클래스 `Animal` 이 주어져 있다.

````python
class Animal:
    def __init__(self, name):
        self.name = name
    def __repr__(self):
        return f"이름: {self.name}"
````

`Animal` 을 상속받아 `bark` 메서드를 가지는 클래스 `Dog` 를 설계하라.

`# 예시`

```python
d = Dog("바둑이")
d
```
```bash
이름: 바둑이
```
```python
d.bark()
```

멍멍!
